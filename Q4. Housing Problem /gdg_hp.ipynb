```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data_train = pd.read_csv('data/hp_train.csv')
data_test = pd.read_csv('data/hp_test.csv')
data_train.shape
data_train.head()
```

```python
from sklearn.model_selection import train_test_split

X_all = data_train.drop(['medv'], axis=1)
y_all = data_train['medv']

num_test = 0.20
X_train, X_test, y_train, y_test = train_test_split(X_all, y_all, test_size=num_test, random_state=23)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lr= LinearRegression()
lr.fit(X_train, y_train)
y_pred= lr.predict(data_test)
```

```python
output = pd.DataFrame({ 'medv': y_pred })
output.to_csv('gdg-housing_problem.csv', index = False)
```
